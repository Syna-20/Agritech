```python
import cv2 
import os 
import glob 
img_dir = "/content/output1" # Enter Directory of all images  
data_path = os.path.join(img_dir,'*g') 
files = glob.glob(data_path) 
data = [] 
for f1 in files: 
    img = cv2.imread(f1) 
    data.append(img)
```

```python
for i in data:
  print(f"{i}.jpg")
```

```python
from imageai.Detection.Custom import CustomObjectDetection
import os

execution_path = os.getcwd()

detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath(detection_model_path="detection_model-ex-028--loss-8.723.h5")
detector.setJsonPath(configuration_json="/content/apple_dataset/json/detection_config.json")
detector.loadModel()
for i in range(9):
    detections = detector.detectObjectsFromImage(input_image="/content/output1/frame_2.jpg", minimum_percentage_probability=60, output_image_path=f"image-new{i}.jpg")
    output_image=cv2.imread(f"image-new{i}.jpg")
    final=output_image.copy()
    count=0
    healthycount=0

for detection in detections:
  if(detection["name"]=='apple'):
    healthycount=healthycount+1
    print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])
    count=count+1
print(healthycount)
print(count)
label=count
# plt.imshow(output_image[:,:,::-1]);plt.title(count)
color = (255, 0, 0)
cv2.putText(final, f"The healthy fruit count is : {count}", (10, 30),cv2.FONT_HERSHEY_PLAIN, 2, color, 4)
cv2_imshow(final)
```
